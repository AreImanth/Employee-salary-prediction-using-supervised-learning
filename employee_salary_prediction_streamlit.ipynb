```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, OneHotEncoder

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

models = {
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}

for name, model in models.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('model', model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    results[name] = acc
    print(f"{name} Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.bar(results.keys(), results.values(), color='skyblue')
plt.ylabel('Accuracy Score')
plt.title('Model Comparison')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Define models
models = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(),
    "KNN": KNeighborsClassifier(),
    "SVM": SVC(),
    "GradientBoosting": GradientBoostingClassifier()
}

results = {}

# Train and evaluate
for name, model in models.items():
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    results[name] = acc
    print(f"{name}: {acc:.4f}")

# Get best model
best_model_name = max(results, key=results.get)
best_model = models[best_model_name]
print(f"\n✅ Best model: {best_model_name} with accuracy {results[best_model_name]:.4f}")

# Save the best model
joblib.dump(best_model, "best_model.pkl")
print("✅ Saved best model as best_model.pkl")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder
import os # For checking if files exist

# --- Configuration and Model/Encoder Loading ---

# Set Streamlit page configuration
st.set_page_config(page_title="Employee Salary Classification", page_icon="💼", layout="centered")

# Define paths for your saved model and encoders
MODEL_PATH = "best_model.pkl"
ENCODERS_DIR = "." # Assuming encoders are in the same directory as app.py

# List of all expected categorical features and their possible categories (MUST match training)
CATEGORICAL_FEATURES_MAP = {
    "education": ["Bachelors", "Masters", "PhD", "HS-grad", "Assoc", "Some-college"],
    "occupation": [
        "Tech-support", "Craft-repair", "Other-service", "Sales",
        "Exec-managerial", "Prof-specialty", "Handlers-cleaners", "Machine-op-inspct",
        "Adm-clerical", "Farming-fishing", "Transport-moving", "Priv-house-serv",
        "Protective-serv", "Armed-Forces"
    ],
    "workclass": ["Private", "Local-gov", "State-gov", "Federal-gov", "Self-emp-not-inc", "Self-emp-inc", "Without-pay", "Never-worked"], # Added "Self-emp-not-inc", "Self-emp-inc", "Without-pay", "Never-worked" as common workclass categories
    "marital_status": ["Married-civ-spouse", "Divorced", "Separated", "Never-married", "Widowed", "Married-spouse-absent", "Married-AF-spouse"], # Added "Married-spouse-absent", "Married-AF-spouse"
    "relationship": ["Husband", "Wife", "Not-in-family", "Unmarried", "Own-child", "Other-relative"], # Added "Own-child", "Other-relative"
    "race": ["White", "Black", "Asian-Pac-Islander", "Amer-Indian-Eskimo", "Other"], # Added "Asian-Pac-Islander", "Amer-Indian-Eskimo"
    "gender": ["Male", "Female"] # Assuming "Others" is not a standard category in the original dataset for gender
}

# Define the exact order of features your model expects (CRITICAL!)
# YOU MUST REPLACE THIS WITH THE ACTUAL LIST OF COLUMN NAMES FROM YOUR X_TRAIN.COLUMNS.TOLIST()
# AFTER ALL PREPROCESSING DURING YOUR MODEL TRAINING.
# This is a placeholder based on the features you provided.
MODEL_FEATURE_NAMES = [
    'age',
    'education',
    'occupation',
    'hours-per-week',
    'experience',
    'workclass',
    'fnlwgt',
    'marital_status',
    'relationship',
    'race',
    'gender',
    'capitalgain',
    'capitalloss'
]

# Load the trained model
try:
    model = joblib.load(MODEL_PATH)
    st.sidebar.success("Model loaded successfully!")
except FileNotFoundError:
    st.sidebar.error(f"Error: Model file '{MODEL_PATH}' not found. Please ensure it's in the correct directory.")
    st.stop()
except Exception as e:
    st.sidebar.error(f"Error loading model: {e}")
    st.stop()

# Load pre-fitted LabelEncoders
# This is crucial for consistent encoding.
label_encoders = {}
for feature, categories in CATEGORICAL_FEATURES_MAP.items():
    encoder_path = os.path.join(ENCODERS_DIR, f"le_{feature}.pkl")
    try:
        label_encoders[feature] = joblib.load(encoder_path)
        # Optional: Verify the loaded encoder's classes_ match expected categories
        # if set(label_encoders[feature].classes_) != set(categories):
        #     st.sidebar.warning(f"Warning: Loaded encoder for '{feature}' has different classes than expected.")
    except FileNotFoundError:
        st.sidebar.warning(
            f"Warning: LabelEncoder for '{feature}' not found at '{encoder_path}'. "
            "Fitting a new encoder on the fly. This is NOT recommended for production "
            "as it can lead to inconsistent encoding if not all categories are present in input. "
            "Please save and load pre-fitted encoders from your training pipeline."
        )
        # Fallback: Fit a new encoder (DANGEROUS FOR PRODUCTION)
        le = LabelEncoder()
        le.fit(categories)
        label_encoders[feature] = le
    except Exception as e:
        st.sidebar.error(f"Error loading LabelEncoder for '{feature}': {e}")
        st.stop()

# --- Preprocessing Function ---

def preprocess_data(df, encoders, feature_names_order):
    """
    Applies Label Encoding and ensures feature order for a DataFrame.
    """
    processed_df = df.copy()

    for feature, le in encoders.items():
        if feature in processed_df.columns:
            # Handle unseen categories during prediction for robustness
            # If a category is not in the encoder's classes, assign a default value
            # (e.g., -1 or the most frequent label from training) or raise an error.
            # For simplicity, we'll try to transform and catch errors.
            try:
                processed_df[feature] = le.transform(processed_df[feature])
            except ValueError as e:
                st.error(f"Error encoding '{feature}': {e}. A category might be present in input that the encoder wasn't trained on.")
                # You might want to assign a default value or drop the row here
                # For now, let's just stop to highlight the issue.
                st.stop()
        else:
            st.warning(f"Feature '{feature}' expected for encoding but not found in input DataFrame.")
            # Depending on your model, you might need to add this column with a default value (e.g., 0 or mode)
            # processed_df[feature] = 0 # Example

    # Ensure all expected features are present and in the correct order
    # Fill any missing columns (e.g., if a new feature was added to MODEL_FEATURE_NAMES
    # but not provided in the input df) with 0.
    processed_df = processed_df.reindex(columns=feature_names_order, fill_value=0)

    return processed_df

# --- Streamlit UI ---

st.title("💼 Employee Salary prediction Model")
st.markdown("Predict whether an employee earns >50K or ≤50K based on input features.")

# Sidebar inputs
st.sidebar.header("Input Employee Details")

# Existing features
age = st.sidebar.slider("Age", 18, 65, 30)
education = st.sidebar.selectbox("Education Level", CATEGORICAL_FEATURES_MAP["education"])
occupation = st.sidebar.selectbox("Job Role", CATEGORICAL_FEATURES_MAP["occupation"])
hours_per_week = st.sidebar.slider("Hours per week", 1, 80, 40)
experience = st.sidebar.slider("Years of Experience", 0, 40, 5)

# New features
workclass = st.sidebar.selectbox("Workclass", CATEGORICAL_FEATURES_MAP["workclass"])
fnlwgt = st.sidebar.slider("Final Weight (fnlwgt)", 100000, 500000, 200000)
marital_status = st.sidebar.selectbox("Marital Status", CATEGORICAL_FEATURES_MAP["marital_status"])
relationship = st.sidebar.selectbox("Relationship", CATEGORICAL_FEATURES_MAP["relationship"])
race = st.sidebar.selectbox("Race", CATEGORICAL_FEATURES_MAP["race"])
gender = st.sidebar.selectbox("Gender", CATEGORICAL_FEATURES_MAP["gender"])
capitalgain = st.sidebar.slider("Capital Gain", 0, 100000, 0)
capitalloss = st.sidebar.slider("Capital Loss", 0, 50000, 0)


# Build input DataFrame from sidebar values
# Ensure column names exactly match those in MODEL_FEATURE_NAMES and your training data
input_data_single = {
    'age': [age],
    'education': [education],
    'occupation': [occupation],
    'hours-per-week': [hours_per_week],
    'experience': [experience],
    'workclass': [workclass],
    'fnlwgt': [fnlwgt],
    'marital_status': [marital_status], # Note: changed 'marital status' to 'marital_status' for consistency
    'relationship': [relationship],
    'race': [race],
    'gender': [gender],
    'capitalgain': [capitalgain],
    'capitalloss': [capitalloss]
}
input_df_raw = pd.DataFrame(input_data_single)

# Preprocess the single input
processed_input_df = preprocess_data(input_df_raw, label_encoders, MODEL_FEATURE_NAMES)

st.write("### 🔎 Input Data (Processed for Model)")
st.write(processed_input_df)

# Predict button
if st.button("Predict Salary Class"):
    if processed_input_df.shape[1] != len(MODEL_FEATURE_NAMES):
        st.error(f"Feature count mismatch! Expected {len(MODEL_FEATURE_NAMES)} features but got {processed_input_df.shape[1]}. Please check your input data and MODEL_FEATURE_NAMES list.")
    else:
        try:
            prediction = model.predict(processed_input_df)
            prediction_proba = model.predict_proba(processed_input_df) # Get probabilities too

            st.success(f"✅ Prediction: {'Income >50K' if prediction[0] == 1 else 'Income <=50K'}")
            #st.info(f"Probability of Income >50K: {prediction_proba[0][1]:.2f}")
            #st.info(f"Probability of Income <=50K: {prediction_proba[0][0]:.2f}")

        except Exception as e:
            st.error(f"An error occurred during prediction: {e}")
            st.write("Please ensure your model expects the exact feature set provided.")


# --- Batch prediction ---
st.markdown("---")
st.markdown("#### 📂 Batch Prediction")
uploaded_file = st.file_uploader("Upload a CSV file for batch prediction", type="csv")

if uploaded_file is not None:
    try:
        batch_data_raw = pd.read_csv(uploaded_file)
        st.write("Uploaded data preview:", batch_data_raw.head())

        # Preprocess the batch data using the same function
        batch_data_processed = preprocess_data(batch_data_raw, label_encoders, MODEL_FEATURE_NAMES)

        if batch_data_processed.shape[1] != len(MODEL_FEATURE_NAMES):
            st.error(f"Feature count mismatch in uploaded CSV! Expected {len(MODEL_FEATURE_NAMES)} features but got {batch_data_processed.shape[1]} after preprocessing. Please check your CSV column names and content.")
        else:
            st.write("Processed batch data preview (for model):", batch_data_processed.head())
            batch_preds = model.predict(batch_data_processed)
            batch_data_raw['Predicted_Salary_Class'] = ['>50K' if p == 1 else '<=50K' for p in batch_preds] # Map back to meaningful labels

            st.write("✅ Predictions:")
            st.dataframe(batch_data_raw.head()) # Show original data with new prediction column

            csv = batch_data_raw.to_csv(index=False).encode('utf-8')
            st.download_button("Download Predictions CSV", csv, file_name='predicted_salary_classes.csv', mime='text/csv')

    except Exception as e:
        st.error(f"An error occurred during batch prediction: {e}")
        st.info("Please ensure your CSV file has the correct column names and data types matching your training data.")

```

```python
from pyngrok import ngrok
import subprocess
import os
import time

# --- Configuration ---
STREAMLIT_PORT = 8501 # Default Streamlit port
NGROK_AUTH_TOKEN = "" # <<< REPLACE THIS WITH YOUR ACTUAL NGROK AUTH TOKEN!

# --- Kill any existing ngrok processes to avoid conflicts ---
print("Killing any existing ngrok processes...")
os.system("killall ngrok")
time.sleep(1) # Give it a moment to terminate

# --- Configure ngrok authtoken (if not already done) ---
try:
    # This writes your authtoken to ngrok's config file
    ngrok.set_auth_token(NGROK_AUTH_TOKEN)
    print("Ngrok authtoken configured successfully.")
except Exception as e:
    print(f"Error configuring ngrok authtoken: {e}")
    print("Please ensure 'YOUR_NGROK_AUTHTOKEN' is replaced with your actual token from ngrok.com.")
    print("You might need to restart the runtime if you're still having issues.")
    exit() # Exit if ngrok cannot be authenticated

# --- Start Streamlit in the background ---
print(f"Starting Streamlit app on port {STREAMLIT_PORT}...")
# We use subprocess.Popen to run Streamlit in the background
# --server.enableCORS true is often needed for ngrok
# --browser.gatherUsageStats false prevents a browser window from trying to open locally
process = subprocess.Popen(
    ["streamlit", "run", "app.py", "--server.port", str(STREAMLIT_PORT), "--server.enableCORS", "true", "--browser.gatherUsageStats", "false"],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE
)

# Give Streamlit a moment to start up
print("Waiting for Streamlit to start...")
time.sleep(5) # Adjust this sleep time if Streamlit takes longer to start

# --- Create ngrok tunnel ---
print("Creating ngrok tunnel...")
try:
    public_url = ngrok.connect(addr=STREAMLIT_PORT, proto="http")
    print(f"Your Streamlit app is available at: {public_url}")
    print("\nIMPORTANT: Keep this cell running to keep the app and tunnel active.")
    print("To stop, interrupt execution (Runtime -> Interrupt execution).\n")

    # Keep the cell running indefinitely to maintain the tunnel
    try:
        while True:
            time.sleep(60) # Keep alive, but check occasionally
            # You could add a simple check here to see if the process is still alive
            if process.poll() is not None:
                print("Streamlit process stopped unexpectedly.")
                break
    except KeyboardInterrupt:
        print("Interrupted by user. Shutting down...")
    finally:
        print("Stopping Streamlit app and ngrok tunnel.")
        process.terminate() # Terminate the Streamlit process
        ngrok.kill() # Kill all ngrok tunnels
        print("Clean up complete.")

except Exception as e:
    print(f"Error creating ngrok tunnel: {e}")
    print("Please ensure your ngrok authtoken is correct and you have internet connectivity.")
    if "Failed to connect to ngrok" in str(e):
        print("This might indicate an issue with your authtoken, firewall, or ngrok servers.")
    process.terminate() # Ensure Streamlit is terminated even on ngrok error
```
